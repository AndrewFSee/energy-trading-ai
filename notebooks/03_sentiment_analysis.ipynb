## 1. Fetch News Articles

## 3. Sentiment Analysis

```python
from src.nlp.sentiment import SentimentAnalyzer
# Note: First run will download the FinBERT model (~400MB)
analyzer = SentimentAnalyzer()

sample_texts = [
    "OPEC cuts production by 1 million barrels per day sending oil prices higher",
    "Crude oil inventories rise more than expected, weighing on WTI prices",
    "Oil prices stable as markets await Fed decision on interest rates",
]

for text in sample_texts:
    result = analyzer.analyse_text(text)
    print(f"Text: {text[:60]}...")
    print(f"  -> {result['label']} (score: {result['score']:.3f})")
    print()
```
