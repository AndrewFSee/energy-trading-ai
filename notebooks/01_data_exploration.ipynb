Exploratory data analysis of energy price, volume, fundamentals, and macro data.

## Setup

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from dotenv import load_dotenv
load_dotenv()

print("Environment ready!")
```

## 1. Load Price Data

```python
from src.data.price_fetcher import PriceFetcher

fetcher = PriceFetcher()
prices = fetcher.fetch(
    symbols=['wti', 'natural_gas', 'brent', 'vix', 'dxy', 'sp500'],
    start='2015-01-01'
)
print(f"Loaded data for: {list(prices.keys())}")
print(f"WTI shape: {prices['wti'].shape}")
prices['wti'].tail()
```

## 2. Price Statistics

```python
wti = prices['wti']
print("WTI Price Statistics:")
print(wti['Close'].describe())
```

## 3. Return Distribution

```python
import numpy as np

log_returns = np.log(wti['Close'] / wti['Close'].shift(1)).dropna()
print(f"Mean daily return: {log_returns.mean():.4f}")
print(f"Std daily return:  {log_returns.std():.4f}")
print(f"Annualised vol:    {log_returns.std() * np.sqrt(252):.2%}")
```

## 4. Correlation Analysis

```python
close_prices = pd.DataFrame({name: df['Close'] for name, df in prices.items()})
correlation = close_prices.pct_change().corr()
print("Return Correlations:")
print(correlation.round(3))
```
