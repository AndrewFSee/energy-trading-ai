## 1. Load Price Data

```python
from src.data.price_fetcher import PriceFetcher
fetcher = PriceFetcher()
wti_df = fetcher.fetch_single('wti', start='2018-01-01')
print(f"Loaded WTI data: {wti_df.shape}")
```

## 2. Technical Features

```python
from src.features.technical import TechnicalFeatures
tf = TechnicalFeatures()
tech_df = tf.add_all(wti_df)
print(f"Features added: {len(tech_df.columns) - 5}")
print("New columns:", [c for c in tech_df.columns if c not in wti_df.columns][:10])
```

## 3. Seasonal Features

```python
from src.features.seasonal import SeasonalFeatures
sf = SeasonalFeatures()
seasonal_df = sf.add_all(tech_df)
print("Seasonal features:", [c for c in seasonal_df.columns if 'season' in c or 'month' in c or 'sin' in c])
```

## 4. Full Pipeline

```python
from src.features.pipeline import FeaturePipeline
pipeline = FeaturePipeline()
feature_df = pipeline.build(price_df=wti_df)
target = pipeline.create_targets(feature_df, horizon=5)
print(f"Feature matrix: {feature_df.shape}")
print(f"Target range: [{target.min():.4f}, {target.max():.4f}]")
```
